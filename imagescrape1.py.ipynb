```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import io
import time
from PIL import Image
PATH = "C:\\Programming Languages\\Python\\web scrapping\\chromedriver.exe"

wd = webdriver.Chrome(PATH)

def get_images(wd, delay, max_imgs):
    def scroll_down(wd):
        wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(delay)
        
    url = "https://www.google.com/search?q=highway%20road%20divider%20images&tbm=isch&hl=en&rlz=1C1VDKB_enIN989IN989&sa=X&ved=0CCEQtI8BKAJqFwoTCKidzaqMpfkCFQAAAAAdAAAAABAG&biw=1476&bih=690#imgrc=M6OnQ2-s51IdHM"
    wd.get(url)
    
    image_urls = set()
    skips = 0


    while len(image_urls)+skips < max_imgs:
        scroll_down(wd)
        thumbnails = wd.find_elements(By.CLASS_NAME,"Q4LuWd")
        for img in thumbnails[len(image_urls)+skips:max_imgs]:
            try:
                img.click();
                time.sleep(delay)
            except:
                continue
            images = wd.find_elements(By.CLASS_NAME, "n3VNCb")
            for image in images:
                if image.get_attribute('src') in image_urls:
                    max_imgs +=1
                    skips +=1
                    break
                if image.get_attribute('src') and 'http' in image.get_attribute('src'):
                    image_urls.add(image.get_attribute('src'))
                    print(f"Found image! {len(image_urls)}")
    return image_urls
    
    
    
def download_img(download_path, url, file_name):
    try: 
        image_content = requests.get(url).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file)
        file_path = download_path + file_name

        with open(file_path,"wb") as f:
            image.save(f, "JPEG")
        print("success")
    except Exception as e:
        print("Failed -", e)
        
        
        
urls  = get_images(wd, 1, 70)
for i, url in enumerate(urls):
    download_img("images/", url, "img" + str(i)".jpg")
wd.quit()
```
